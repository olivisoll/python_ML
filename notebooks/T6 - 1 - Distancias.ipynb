# Distancias

```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
data = pd.read_csv("../datasets/movies/movies.csv", sep=";")
```

```python
data
```

```python
movies = data.columns.values.tolist()[1:] #eliminamos la primera columna que es el user_id
```

```python
movies
```

```python
# Distancia Manhattan
dd1 = distance_matrix(data[movies], data[movies], p=1)
# Distancia Eucl√≠dea
dd2 = distance_matrix(data[movies], data[movies], p=2)
# Distancia 10
dd10 = distance_matrix(data[movies], data[movies], p=10)
```

```python
def dm_to_df(dd, col_name):
    import pandas as pd
    return pd.DataFrame(dd, index=col_name, columns=col_name)
```

```python
dm_to_df(dd1, data["user_id"])
```

```python
dm_to_df(dd2, data["user_id"])
```

```python
dm_to_df(dd10, data["user_id"])
```

```python
# Al aumentar la p (√≠ndice de Minkowsky), la distancia disminuye
```

## Clustering jer√°rquico con herramientas de Python

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], "ward")
Z
```

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma jer√°rquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=11.0)
plt.show()
```

```python
Z = linkage(data[movies], "average")
plt.figure(figsize=(25,10))
plt.title("Dendrograma jer√°rquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=11.0)
plt.show()
```

```python
Z = linkage(data[movies], "complete")
plt.figure(figsize=(25,10))
plt.title("Dendrograma jer√°rquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=11.0)
plt.show()
```

```python
Z = linkage(data[movies], "single")
plt.figure(figsize=(25,10))
plt.title("Dendrograma jer√°rquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=11.0)
plt.show()
```

```python
Z = linkage(data[movies], "weighted")
plt.figure(figsize=(25,10))
plt.title("Dendrograma jer√°rquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=11.0)
plt.show()
```

```python
Z = linkage(data[movies], "single", metric="")
plt.figure(figsize=(25,10))
plt.title("Dendrograma jer√°rquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=11.0)
plt.show()
```
