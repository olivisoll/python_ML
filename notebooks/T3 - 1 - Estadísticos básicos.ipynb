## Medidas de centralización

Nos sirven para ver cómo se sitúan los datos.

* **Media aritmética**
* **Mediana**: si ordenaramos los datos de un array de menor a mayor, la mediana es el valor central (o los 2 valores centrales en caso de que la muestra sea par).
* **Moda**: valor que más aparece
* **Percentiles**: lo mismo que la mediana pero para cualquier valor entre 0 y 1.

Generamos una lista rándom de números:

```python
# Generar números aleatorios
import random
random.seed(0)
A = (random.sample(range(100000), 60))
A
```

```python
mean = np.mean(A) # Media aritmética
mean
```

```python
np.median(A) # Mediana
```

```python
from scipy import stats # Librería para el cálculo de la moda
stats.mode(A)
```

```python
# percentil 100=p1 0=p0, 0.25 1er cuantil 0.75 3er cuantil
np.percentile(A, 25) # return 0.25 percentile, e.g. 1er cuantil.
```

## Medidas de dispersión

La varianza y desviación típica, nos indican si los valores de nuestra muestra se desplazan mucho o poco con respecto de la media.

* **Varianza**: medir cómo los valores se desvían de la media. Siempre será un valor positivo.
* **Desviación típica**: raíz cuadrada de la varianza.
* **Coeficiente de variación**: mide la variabilidad relativa entre la media y la desviación típica (es un %).

```python
np.var(A) # Varianza
```

```python
std = np.std(A) # Desviación típica
std
```

```python
# Coeficiente de variación = std/mean*100
# Variabilidad relativa entre la media y la std, si hay mucha variabilidad será grande el coeficiente
std/mean*100
```

## Medidas de asimetría

Momento de orden r respecto de la media. El momento de orden r son los momentos de distribución respecto de la media.

```python
from IPython.display import Image
Image(filename="resources/momento_orden_r.png")
```

#### Asimetría de Fisher

Momento de orden 3 dividido la desviación típica al cubo. 

* Si este **valor es 0**, nuestra función es perfectamente simétrica (esto es raro), se distribuye igual, por ejemplo la distribución normal.
* Si el **coeficiente es > 0**, significa que cuánto más positivo es este valor, más desplazada está la distribución hacia la izquierda, de modo que tenemos una asimetría positiva y la media queda muy por encima de la distribución.
* Si el **coeficiente es < 0**, significa que cuánto más negativo es este valor más desplazada está la distribución hacia la derecha, de modo que tenemos una asimetría negativa y la media queda muy por debajo de la distribución de valores.

```python
Image(filename="resources/asimetria_fisher.png")
```

```python
# 3rd moment: Asimetría de Fisher
import scipy.stats as sp # para modelos de asimetría
asimetria = sp.skew(A)
asimetria
```

#### Coeficiente de curtosis

Momento de orden 4 dividido por la desviación típica a la cuarta menos 3.

* Si este **coeficiente es 0**, se asemeja a la distribución normal en forma, no en valores. Está compensado tanto en el centro como en las colas.
* **Coeficiente > 0**, se dice que es **leptocúrtica**, es decir, sus valores se concentran mucho en el centro y las colas están degradadas.
* **Coeficiente < 0**, se dice que es una distribución **platicúrtica**, hay pocos valores que se concentren respecto a la media y muchos en las colas.

```python
Image(filename="resources/curtosis.png")
```

```python
# 4th Moment: Curtosis
curtosis = sp.kurtosis(A)
curtosis
```
