# Implementando k-means con Python

```python
import numpy as np
```

```python
data = np.random.random(90).reshape(30,3)
data
```

```python
# Generamos 2 baricentros aleatorios, y los clusters a partir de ellos

c1 = np.random.choice(range(len(data)))
c2 = np.random.choice(range(len(data)))
clust_centers = np.vstack([data[c1], data[c2]])
clust_centers
```

```python
from scipy.cluster.vq import vq
```

```python
vq(data, clust_centers) 
# el primer array nos muestra a qué cluster pertenece cada elemento
# el segundo array nos da la distancia de cada una de las 30 observaciones (elementos) 
# al baricentro del cluster al que pertenece
```

```python
from scipy.cluster.vq import kmeans
```

```python
# visualización de los baricentros de los clusters, el número final es la suma de los cuadrados de los errores
kmeans(data, clust_centers)
```

```python
# a k-means le podemos indicar que cree los clusters, diciéndole cuántos queremos en vez de los baricentros
kmeans(data, 2)
```
