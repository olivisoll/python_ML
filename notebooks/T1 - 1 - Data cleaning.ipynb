### Carga de datos a través de la función open

Función de python, lee el dataset línea a línea. Para hacer cargas manuales de los datos.

````python
data3 = open(mainpath + "/" + "customer-churn-model/Customer Churn Model.txt", "r")
````

````python
cols = data3.readline().strip().split(",")
n_cols = len(cols)
````

````python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []

for line in data3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1

print ("El data set tiene %d filas y %d columnas" % (counter, n_cols))
````

````python
df3 = pd.DataFrame(main_dict)
df3.head()
````

### Lectura y escritura de ficheros

````python
infile = mainpath + "/" + "customer-churn-model/Customer Churn Model.txt"
outfile = mainpath + "/" + "customer-churn-model/Tab Customer Churn Model.txt"
````

````python
with open(infile, "r") as infile1:
    with open(outfile, "w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
````

````python
df4 = pd.read_csv(outfile, sep="\t")
df4.head()
````

### Ejercicio
Ejercicio de carga de un csv usando la librería urllib3

````python
import urllib3
http = urllib3.PoolManager()
r = http.request('GET', medals_url)
response = r.data

# Decodificamos la respuesta, ya que viene en formato string binario
str_data = response.decode('utf-8')

# Partimos el string para recuperar las líneas
lines = str_data.split("\n")

# La primera línea es la cabecera, la partimos por comas para saber cuántas columnas tendremos
col_names = lines[0].split(",")
n_cols = len(col_names)
counter = 0
main_dict = {}
for col in col_names:
    main_dict[col] = []

for line in lines:
    #Hay que saltarse la primera línea puesto que contiene la cabecera
    if counter > 0:
        values = line.strip().split(",")
        for i in range(len(col_names)):
            main_dict[col_names[i]].append(values[i])
    counter += 1

print ("El data set tiene %d filas y %d columnas" % (counter, n_cols))
medals_df = pd.DataFrame(main_dict)
medals_df.head()

# Elegimos dónde guardarlo (en la carpeta athletes)
mainpath = "/Users/olivialuis/AnacondaProjects/CursoPythonML/datasets"
filename = "athletes/downloaded-medals-olivia."
fullpath = os.path.join(mainpath, filename)

medals_df.to_csv(fullpath + "csv")
medals_df.to_json(fullpath + "json")
medals_df.to_excel(fullpath + "xls")
````
