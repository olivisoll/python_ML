# Clustering jerárquico y dendrogramas

Antes de empezar, pongamos un poco de notación para hablar todos el mismo idioma
* X dataset (array de n x m) de puntos a clusterizar
* n número de datos
* m número de rasgos
* Z array de enlace del cluster con la información de las uniones
* k número de clusters

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
np.random.seed(4711)
a = np.random.multivariate_normal([10,0],[[3,1],[1,4]], size = [100,])
b = np.random.multivariate_normal([0,20],[[3,1],[1,4]], size = [50,])
X = np.concatenate((a,b))
print(X.shape)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
Z = linkage(X, "ward") # creamos matriz de enlace, linkage por defecto coge la distancia euclídea
```

### Truncar el dendrograma

```python
plt.figure(figsize=(25,10))
plt.title("Dendrograma del clustering jerárquico")
plt.xlabel("Índices de la muestra")
plt.ylabel("Distancias")
dendrogram(Z, leaf_rotation=90., leaf_font_size=12.0, truncate_mode="lastp", 
           p=12, # p=10 (últimos 10 clusteres unidos)
           show_leaf_counts=False, # mostrar u ocultar los elementos de los nodos hoja
           show_contracted=True, #mostrar la distribución contraída en las ramas, para evitar sobrecargar el dendrograma
          ) 
plt.show()
```

### Dendrograma personalizado

```python
def dendrogram_tuned(*args, **kwargs):
    #plt.figure(figsize=(25,10))
    max_d = kwargs.pop("max_d", None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold'] = max_d
    
    annotate_above = kwargs.pop('annotate_above', 0)
    
    ddata = dendrogram(*args, **kwargs)
    
    if not kwargs.get('no_plot', False):
        plt.title("Clustering jerárquico con dendrograma truncado")
        plt.xlabel("Índice del dataset (o tamaño del cluster)")
        plt.ylabel("Distancia")
        for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            if y>annotate_above:
                plt.plot(x,y, 'o', c=c)
                plt.annotate('%.3g'%y, (x,y), xytext = (0,-5), textcoords="offset points", va="top", ha="center")

    if max_d:
        plt.axhline(y=max_d, c='k')
        
    return ddata
            
```

```python
dendrogram_tuned(Z, truncate_mode='lastp', p=12, leaf_rotation=90., leaf_font_size=15., 
                 show_contracted=True, annotate_above=10, max_d=20)
plt.show()
```

```python
depth = 5
incons = inconsistent(Z, depth)
incons[-10:]
```

### Método del codo

```python
last = Z[-10:, 2]
last_rev = last[::-1]
print(last_rev)
idx = np.arange(1, len(last) + 1)
plt.plot(idx, last_rev)

acc = np.diff(last, 2)
acc_rev = acc[::-1]
plt.plot(idx[:-2]+1, acc_rev)
plt.show()
k = acc_rev.argmax() + 2
print("El número óptimo de clusteres es %s"%str(k))
```

```python
c = np.random.multivariate_normal([40,40], [[20,1],[1,30]], size=[200,])
d = np.random.multivariate_normal([80,80], [[30,1],[1,30]], size=[200,])
e = np.random.multivariate_normal([0,100], [[100,1],[1,100]], size=[200,])
X2 = np.concatenate((X,c,d,e),)
plt.scatter(X2[:,0], X2[:,1])
plt.show()
```

```python
Z2 = linkage(X2, "ward")
plt.figure(figsize=(10, 10))
dendrogram_tuned(
    Z2,
    truncate_mode="lastp",
    p=30,
    leaf_rotation=90.,
    leaf_font_size=10.,
    show_contracted=True,
    annotate_above = 40,
    max_d = 170
)
plt.show()
```

## Recuperar los clusters y sus elementos

```python
from scipy.cluster.hierarchy import fcluster
```

```python
max_d = 20 # altura a la que queremos cortar
clusters = fcluster(Z, max_d, criterion="distance")
clusters # un elemento por cada dato, indicando en qué cluster debe ir cada elemento
```

```python
k=2 # número de clusteres que queremos
clusters = fcluster(Z,k,criterion="maxclust")
clusters
```

```python
fcluster(Z, 8, depth=10) # por defecto, el criterio que usa es el método de inconsistencia para elaborar las distancias
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X[:,0],X[:,1], c= clusters, cmap="prism")
plt.show()
```
